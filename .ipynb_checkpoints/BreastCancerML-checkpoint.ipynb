```python
#verileri bağımsız (X) ve bağımlı (Y) veri olarak split ediyoruz.Yani bölüyoruz
X=df_2.iloc[:,2:31].values #Tahmin etmek için kullandığımız Train verilerimiz
Y=df_2.iloc[:,1].values  #Tahmin etmeye çalıştımız değerlerdir target 

```

```python
#%75 train ve %25 test olarak verimizi Split ediyoruz
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
#X tarin veriilerimi Y target yani tahmin edeceğimiz veriler
```

```python
#verimize Feature Scaling yapıyoruz yani özelliklerimizi 
#belli Scaller metodlarala ölçelendirmiş olacağız
from sklearn.preprocessing import StandardScaler
#standartScaler ortalamanın 0 standart sapmanın 1 olduğu değerlere göre değer aralığı belirler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
X_train
```

```python
#Modelimiz için bir fonksiyon oluşturuyoruz ve kullanacağımız 
#algoritmaları burada tanımlıyoruz
#fonksiyonumuzn parametrekleri X_train ve Y_train oluyor.
def models(X_train,Y_train):
    #Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion="entropy",random_state=0)
    tree.fit(X_train,Y_train)
    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=0)
    forest.fit(X_train,Y_train)
    #SVM Classsifier(SVC)
    from sklearn.svm import SVC
    svc=SVC(kernel="linear",C=1)
    svc.fit(X_train,Y_train)
    #KNN Classifier
    from sklearn.neighbors import KNeighborsClassifier
    knn=KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train,Y_train)
    
    #Training verilerinin doğruluk oranlarını yazdıralım
    print("[0]Logistic Regression Training Accuracy:",log.score(X_train,Y_train))
    print("[1]Decision Tree Training Accuracy:",tree.score(X_train,Y_train))
    
    print("[2]Random Forest Classifier Training Accuracy:",forest.score(X_train,Y_train))
    
    print("[3]SVM Classsifier(SVC) Training Accuracy:",svc.score(X_train,Y_train))
    
    print("[4]KNN Classifier Training Accuracy:",knn.score(X_train,Y_train))
    
    return log,tree,forest,svc,knn
    
    
```

```python
#models fonksiyonunu çağırıyoruz
model=models(X_train,Y_train)
```

```python
#Test verilerimiizn doğruluk oranlarını görmek içn conffusion matrix bakıyoruz
from sklearn.metrics import confusion_matrix
for i in range(len (model)):
    print("Algoritma",i)
    cm=confusion_matrix(Y_test,model[i].predict(X_test))
    #model fonk.dan gelen algoritmalara göre değerleri alıyoruz
    #Y_test doğru tahminler, predict(X_test) modelin tahmini
    #bu ikisinin oranı bize test verilerinin doğruluğunu verir
    TP=cm[0][0]
    TN=cm[1][1]
    FN=cm[1][0]
    FP=cm[0][1]
    print(cm)
    print("Testing Accuracy=",(TP+TN)/(TP+TN+FN+FP))
    print("\n")
   # print("Algoritma:",model[i])
```

```python
#Modelimizin metriklerini görelim
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
    print("Algoritma:",i)
    print(classification_report(Y_test,model[i].predict(X_test)))
    print(accuracy_score(Y_test,model[i].predict(X_test)))
    print("\n")
```
